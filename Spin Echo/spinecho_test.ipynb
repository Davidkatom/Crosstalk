```python
from Spinecho_exp import SpinEcho
import numpy as np
import matplotlib.pyplot as plt
```

```python
import random

n = 2
W = [random.gauss(0, 3) for _ in range(n)]
L = [random.gauss(3, 2) for _ in range(n)]
J = [random.gauss(0, 3) for _ in range(n - 1)]
shots = 10000
W = [1,1]
j = [2]
exp = []
tau = np.linspace(0, 10, 50)
for t in tau:
    echo = SpinEcho(n, t, shots, J, W, L)
    echo.create_echo_sequence(t,1)
    exp.append(echo)
```

```python
Z = [exp.zi[0] for exp in exp]
plt.plot(tau+1, Z)
```

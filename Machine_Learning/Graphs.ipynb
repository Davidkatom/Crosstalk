```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [4, 3]  # width, height
plt.rcParams["figure.dpi"] = 150


def super_plot(xl='x', yl='f(x)', title='', base_size=16, xtickx_size=16):
    # base_size =20
    plt.xlabel(xl, fontsize=base_size + 4)
    plt.ylabel(yl, fontsize=base_size + 4)
    plt.title(title, fontsize=base_size + 3)
    plt.legend(fontsize=base_size - 2, frameon=False)
    plt.xticks(10, fontsize=xtickx_size + 4)
    plt.tick_params(axis='both', which='major', labelsize=xtickx_size)
    plt.show()
```

```python
#2 qubits - shots
Shots_2q = [100, 1000, 10000, 100000, 1000000]
Errors_2q = [50.530481, 18.415648, 7.961954, 6.401243, 6.696693]
plt.plot(np.array(Shots_2q), Errors_2q, linestyle='--', marker='o', color='blue', markersize=5)
# plt.title(f'Shots={Decay} ')
plt.xlabel('Shots')
plt.ylabel('Median error (%)')
plt.grid()
# plt.yscale('log')
plt.xscale('log')

plt.rcParams['lines.markersize'] = 10
# Example size for markers
plt.rcParams['font.size'] = 12  # Example size for font
plt.title('2 qubits - shots')

plt.grid()
plt.legend()
```

```python
%matplotlib inline

import panel as pn
import matplotlib.pyplot as plt
from io import BytesIO

# Enable panel in Jupyter Notebook (or Jupyter-compatible environments)
pn.extension()

# Define a function that generates a plot based on the selection
def plot_selection(option):
    plt.figure(figsize=(6, 4))
    plt.plot([1, 2, 3], [1, 2, 3], label=option)
    plt.legend()
    plt.title(f'Selected option: {option}')

    buf = BytesIO()
    plt.savefig(buf, format='png')
    buf.seek(0)
    plt.close()
    return pn.pane.Image(buf, format='png')

# Create a dropdown widget
dropdown = pn.widgets.Select(name='Option', options=['None', 'Option 1', 'Option 2', 'Option 3'])

# Set up reactive plot display
interactive_plot = pn.bind(plot_selection, option=dropdown)

# Display the dropdown and plot together
display = pn.Column(dropdown, interactive_plot)
display.show()  # Use .show() to display in PyCharm, or .servable() in Jupyter
```

```python
import os
import glob
import pandas as pd
import plotly.graph_objects as go
import ipywidgets as widgets


folder_path = 'C:\\Projects\\Crosstalk\\data_for_paper\\'
all_files = os.listdir(folder_path)
csv_files = [f for f in all_files if f.endswith('.xlsx') and os.path.isfile(os.path.join(folder_path, f))]
# print(csv_files)
# Create a dropdown widget with the CSV file names
dropdown = widgets.Dropdown(
    options=csv_files,
    value=None,
    description='Select CSV file:',
)

def plot_error_bars(filename):
    if filename is not None:
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)

        # Ensure there is at least one column
        if df.shape[1] < 3:
            print("The CSV file does not contain enough columns to plot error bars.")
            return

        # The first column is the X-axis
        x_col = df.columns[0]
        x_data = df[x_col].values

        # Remaining columns are in pairs: median and std
        remaining_cols = df.columns[1:]

        # Check if the number of remaining columns is even
        if len(remaining_cols) % 2 != 0:
            print("The CSV file columns after the first should be in median-std pairs.")
            return

        # Group the remaining columns into median-std pairs
        pairs = [(remaining_cols[i], remaining_cols[i+1]) for i in range(0, len(remaining_cols), 2)]

        # Create the figure
        fig = go.Figure()

        for median_col, std_col in pairs:
            if median_col not in df.columns or std_col not in df.columns:
                print(f"Columns {median_col} and {std_col} are missing in the DataFrame.")
                continue

            y_data = df[median_col].values
            y_error = df[std_col].values

            # Add error bars to the figure
            fig.add_trace(go.Scatter(
                x=x_data,
                y=y_data,
                mode='lines+markers',
                name=median_col.replace('_median', ''),
                error_y=dict(type='data', array=y_error, visible=True)
            ))

        fig.update_layout(
            title=f"Error Bars Plot for {filename}",
            xaxis_title=x_col,
            yaxis_title='Values',
            legend_title='Datasets',
            width=800,
            height=600
        )

        fig.show()

# Use the interactive function to update the plot based on the selected CSV file
widgets.interactive(plot_error_bars, filename=dropdown)
```
